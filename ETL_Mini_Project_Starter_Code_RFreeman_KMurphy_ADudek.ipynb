### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
columns = crowdfunding_info_df.columns
print(columns)
```

```python
# Assign the category and subcategory values to category and subcategory columns. crowdfunding_info_df['category & sub-category'].str.split('&', n=1,expand= True)
split_values = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)

# Assign the values to new columns
crowdfunding_info_df['category']=split_values[0]
crowdfunding_info_df['subcategory'] = split_values[1]

# Disply updated dataframe
crowdfunding_info_df.head()
```

```python
# Get the unique categories and subcategories in separate lists.
categories = crowdfunding_info_df['category'].unique()
subcategories = crowdfunding_info_df['subcategory'].unique()

print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
dv_categories = crowdfunding_info_df['category'].nunique()
dv_subcategories = crowdfunding_info_df['subcategory'].nunique()

print(dv_categories)
print(dv_subcategories)
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 
cat_ids = ["cat" + str(category_id) for category_id in category_ids]

print(cat_ids)
```

```python
# Use a list comprehension to add "subcat" to each subcategory_id. 
scat_ids = ["subcat" + str(subcategory_id) for subcategory_id in subcategory_ids]


print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
category_df = pd.DataFrame({'category_id': cat_ids,'category': categories})

# Display new category dataframe with category ids
category_df
```

```python
# Create a category DataFrame with the subcategory_id array as the subcategory_id and subcategories list as the subcategory name. 
subcategory_df = pd.DataFrame({'subcategory_id': scat_ids,'subcategory': subcategories})

# Display new subcategory dataframe with subcategory ids
subcategory_df
```

```python
# Export categories_df and subcategories_df as CSV files.
category_df.to_csv("Database_CSV/category.csv", index=False)

subcategory_df.to_csv("Database_CSV/subcategory.csv", index=False)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
campaign_df.rename(columns={'blurb':"description",
                           "launched_at":"launched_date",
                           "deadline":"end_date"}, inplace=True)

campaign_df
```

```python
# Convert the goal and pledged columns to a `float` data type.
campaign_df["goal"] = campaign_df["goal"].astype('float')
campaign_df["pledged"] = campaign_df["pledged"].astype('float')

campaign_df
```

```python
# Check the datatypes
campaign_df.dtypes
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt
campaign_df["launched_date"] = pd.to_datetime(campaign_df["launched_date"], unit='s').dt.strftime('%Y-%m-%d')
campaign_df["end_date"] = pd.to_datetime(campaign_df["end_date"], unit='s').dt.strftime('%Y-%m-%d')

campaign_df
```

```python
# Merge the campaign_df with the category_df on the "category" column and 
# the subcategory_df on the "subcategory" column.
campaign_merged = campaign_df.merge(category_df, on='category', how='inner')
campaign_merged_df = campaign_merged.merge(subcategory_df, on='subcategory', how='inner')

campaign_merged_df.tail(10)
```

```python
# Drop unwanted columns

unwanted_columns = ['staff_pick','spotlight','category & sub-category','category','subcategory']
campaign_merged_df = campaign_merged_df.drop(columns=unwanted_columns)

campaign_merged_df

campaign_cleaned = campaign_merged_df

campaign_cleaned
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned.to_csv("Database_CSV/campaign.csv", index=False)
```

### Option 1: Use Pandas to create the contacts DataFrame.

```python
# Iterate through the contact_info_df and convert each row to a dictionary.
import json
import pandas as pd
dict_values = []



# Provide the correct file path to your Excel file as a string
excel_file = "contacts .xlsx"

# Read the Excel file into a DataFrame
contact_info_df = pd.read_excel(excel_file)


# Create the contacts DataFrame
contacts = pd.DataFrame(contact_info_df)

# Iterate through the DataFrame
for _, row in contacts.iterrows():
    # Process each row of data
    dict_values = row.to_dict()
    

# Print out the list of values for each row.
print(dict_values)
```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.

import pandas as pd

# Read the Excel file into a DataFrame
contact_info_df = pd.read_excel(excel_file)

# Create an empty DataFrame for contact_info
contact_info = pd.DataFrame(columns=['contact_id', 'name', 'email'])

# Iterate through the DataFrame
for _, row in contact_info_df.iterrows():
    # Extract the necessary values from the row
    contact_id = row['contact_id']
    name = row['name']
    email = row['email']
    
     # Remove the trailing '}' character from the email
    email = email.rstrip('}')
    email = email.rstrip('""')
    

    # Create a new DataFrame row with the extracted values
    contact_info_row = pd.DataFrame([[contact_id, name, email]], columns=['contact_id', 'name', 'email'])

    # Concatenate the new row to the contact_info DataFrame
    contact_info = pd.concat([contact_info, contact_info_row], ignore_index=True)


# Print the resulting contact_info DataFrame
print(contact_info.to_string(index=False))

```

```python
# Check the datatypes.

contact_info['contact_id'] = contact_info['contact_id'].astype('int64')

print(contact_info.info())
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 


# Drop the contact_name column

import pandas as pd


# Split the "name" column into "first_name" and "last_name" columns
contact_info[['first_name', 'last_name']] = contact_info['name'].str.split(expand=True)


# Remove double quotes from the "first_name", "last_name", and "email" columns
contact_info['first_name'] = contact_info['first_name'].str.replace('"', '')
contact_info['last_name'] = contact_info['last_name'].str.replace('"', '')
contact_info['email'] = contact_info['email'].str.replace('"', '')


# Drop the original name column
contact_info.drop('name', axis=1, inplace=True)

# Print the updated DataFrame
print(contact_info.to_string(index=False))
```

```python
# Reorder the columns

contact_info = contact_info[['contact_id', 'first_name', 'last_name', 'email']]

# Print the updated DataFrame
print(contact_info.to_string(index=False))
```

```python
# Check the datatypes one more time before exporting as CSV file.

contact_info['contact_id'] = contact_info['contact_id'].astype('int64')

print(contact_info.info())
```
